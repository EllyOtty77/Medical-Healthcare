```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import pandas as pd

r = requests.get('https://www.healthywa.wa.gov.au/Health-conditions/Health-conditions-A-to-Z')
soup = BeautifulSoup(r.text, 'lxml')
```

```python
main = soup.find('div', {'class': 'az-results'})
entries  = main.find_all('li')

diseases = [entry.text.strip() for entry in entries]

links = [entry.find('a')['href'] for entry in entries]
urls = [f'https://www.healthywa.wa.gov.au/{link}' for link in links]
```

```python
# Create a DataFrame with diseases and URLs
data = {'Disease': diseases, 'URL': urls}
df = pd.DataFrame(data)

# Sort the DataFrame alphabetically by disease name
df_sorted = df.sort_values(by='Disease')

# Drop duplicate rows based on both 'Disease' and 'URL'
df_unique = df_sorted.drop_duplicates()

# Reset the index
df_unique.reset_index(drop=True, inplace=True)
```

```python
df_unique.to_csv('medical_conditions.csv', index=False)

# Read the saved CSV file
pd.read_csv('medical_conditions.csv')
```
